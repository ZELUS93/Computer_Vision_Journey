# Changing Color-Space 

## Goal 
The objective of this tutorial is to learn how to read and convert an image from one color-space to another (from BGR to HSV) using Python and OpenCV.  

The code shown below was tested using Python 3.7.9 and version 3.4.2 of OpenCV

## Introduction 
HSV is the description of hue,saturation, and value(brigthness/luminance) and was developed to be more "intuitive" in manipulating with color and also to approximate the way humans perceive and interpret colors.

Hue defines the color itself. Saturation indicates the degree to which the hue differs from a neutral gray. Value indicates the illumination level. 

## It's time to get our hands dirty ! 
  

```python
# The first thing we are going to do is to import the necessary packages/libraries we will need for this tutorial
# Importing the cv2 module. This module will make available the functions we need to both read the image and convert it to another color space:
import cv2 
# Importing the matplotlib.pyplot for viewing the images: 
from matplotlib import pyplot as plt  

# Now we are ready to load and examine an image
# To read the original image, we simply need to call the "imread" function. As an input of this function we need to pass a string with the path to the image file we want to read. The image is saved under the name "image":
image  = cv2.imread("C:\\Users\\El Mehdi Naor\\Jupyter\\trex.png")  
image2 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
# A little check to see if the image loaded has the right color:
row, col = 1,2 
fig, axs = plt.subplots(row,col,figsize=(10,10))
fig.tight_layout()

axs[0].imshow(image)
axs[0].set_title('BGR')

axs[1].imshow(image2)
axs[1].set_title('RGB')

plt.show()
print('Images displayed') 
# The first image loaded seems in the wrong color precisely because it is loaded as "BGR" but "plt.show" thinks it is in RGB. To convert this to the proper color space, we use the "cv2.cvtCOLOR" method
# To convert to HSV, we should use the "COLOR_RGB2HSV" flag of the "cv2.cvtCOLOR" method. The converted image is saved under the name "HSV_image":
HSV_image = cv2.cvtColor(image2,cv2.COLOR_RGB2HSV)

# Now let's see what our image looks like: 
plt.figure(figsize=(10,10))
plt.imshow(HSV_image)
print('HSV_image displayed') 

# Let's split the image in channels and see what each one looks like:
row, col = 2,2 
fg, axs  = plt.subplots(row,col, figsize=(15,10))

axs[0][0].imshow(HSV_image,)
axs[0][0].set_title('HSV Image')

axs[0][1].imshow(HSV_image[:,:,0], cmap="gray")
axs[0][1].set_title('Hue Channel')

axs[1][0].imshow(HSV_image[:,:,1], cmap="gray")
axs[1][0].set_title('Saturation Channel')

axs[1][1].imshow(HSV_image[:,:,2], cmap="gray")
axs[1][1].set_title('Value Channel') 

plt.show()
print('Images Displayed')
```

```python
# The last step, which is saving the images in the working directory using the "cv2.imwrite(filename,image)" method
# The first parameter of the image reprents the file name. The filename must include image format like ".jpg" and ".png"
# The second parameter is the image that is to be saved 
# This method returns true if image is saved successfully

# Save HSV_image:
cv2.imwrite('HSV_image.png',HSV_image)

# Save Hue Channel image: 
cv2.imwrite('Hue_Channel.png',HSV_image[:,:,0])

# Save Saturation Channel image: 
cv2.imwrite('Saturation_Channel.png',HSV_image[:,:,1])

# Save Value Channel image: 
cv2.imwrite('Value_Channel.png',HSV_image[:,:,2])
```
