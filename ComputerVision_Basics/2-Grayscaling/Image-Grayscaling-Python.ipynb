# Grayscaling 
## Grayscaling is a process in which an image is converted from a full color to shades of grey (black and white)
In OpenCV, many functions grayscale images before processing. This is done because it simplifies the image, acting almost as a noise reduction and increasing processing time as there is less information in the image.

## Let's convert our color image to greyscale
The first thing to always do is to import the packages/libraries we will need for this example:

```python
import cv2 
import numpy as np 
from matplotlib import pyplot as plt 
print("Packages/Libraries successfully imported !")
```

```python
# Load our Image: 
image = cv2.imread('C:\\Users\\El Mehdi Naor\\Jupyter\\trex.png')

# We use the 'cvtColor(...,...)', to convert the colored image into a grayscaled one
# As first input, this function receives the original image. As second input, it receives the color space conversion code 
# Since we want to convert our original image from the BGR color space to gray, we use the code COLOR_2BGR2GRAY:
grayImage = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

# Display the results:
row, col = 1,2 
fig, axs = plt.subplots(row,col, figsize=(10,10))
fig.tight_layout() 

# X-label of the colored image: 
axs[0].imshow(image)
axs[0].set_title('Original')  

#X-label of the grayscaled image: 
axs[1].imshow(grayImage)
axs[1].set_title('Grayscale')

# Plot the results: 
plt.show()
print('Images successfully Plotted !')
```

```python
# Saving the grayscale image using 'imwrite(...,...)': 
cv2.imwrite('Grayscale_Trex.png',grayImage)
```
