First, let's create a file 'named load_display_save.py'to contain our code. Now we can start writing code. The first thing we are going to do is import the packages/libraries we will need for this example:

```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
print("Packages/Libraries successfully imported")
```

Let's now load our first image:

```python
#Load an image using 'imread' specifying the path of the image
image=cv2.imread("C:\\Users\\El Mehdi Naor\\Jupyter\\images\\trex.png")

#Our file 'trex.png' is now loaded and stored in a variable we named 'image'
#To display our image, we use "plt.imshow(...)"
plt.title("TREX") 
plt.imshow(image)

#Hides the graph ticks and x/y axis
plt.xticks([]), plt.yticks([]) 
plt.show()
print("Image successfully Plotted")
```

The last thing we are going to do is write our image to file in JPG format:

```python
# All we are doing here is providing the path to the file (the first argument) and 
# then the image we want to save (the second argument). It's that simple!
cv2.imwrite("newimage.jpg", image) 
print("The image was successfully saved")
```
